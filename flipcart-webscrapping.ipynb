```python
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uReq
```

```python
flipcart_link = "https://www.flipkart.com/search?q="+input()
```

```python
urlclient = uReq(flipcart_link)
```

```python
urlclient
```

```python
flipcart_page=urlclient.read()
```

```python
flipcart_html=bs(flipcart_page,'html.parser')
```

```python
# flipcart_html
```

```python
bigbox=flipcart_html.findAll("div",{"class":"_1AtVbE col-12-12"})
```

```python
bigbox
```

```python
del(bigbox[:2],bigbox[25:])
```

```python
size = len(bigbox)
print(size)
```

```python
url = []
for i in range(0,size-4):
    product_link="https://www.flipkart.com" + bigbox[i].div.a['href']
    url.append(product_link)
print(url)
```

```python
metaData = []
templs = []
for i in url:
    product_req = requests.get(i)
    product_html = bs(product_req.text,'html.parser')
    title=product_html.find("span",{"class":"B_NuCI"})
    rate=product_html.find_all("div",{"class":"_2d4LTz"})
    numOfRate=product_html.find_all("div",{"class":"row _2afbiS"})
    price=product_html.find_all("div",{"class":"_30jeq3 _16Jk6d"})
    #Title---------------------------->
    textTitle = f'{title}'
    soup = bs(textTitle, 'html.parser')
    textTitle = soup.get_text()
    textTitle = textTitle.replace('[', '').replace(']', '')
    #Rate------------------------------>
    textRate = f'{rate}'
    soup = bs(textRate, 'html.parser')
    textRate = soup.get_text()
    textRate = textRate.replace('[', '').replace(']', '')
    #Comment--------------------------->
    textNOR = f'{numOfRate}'
    soup = bs(textNOR, 'html.parser')
    textNOR = soup.get_text()
    textNOR = textNOR.replace('[', '').replace(']', '')
    #Name------------------------------>
    textPrice = f'{price}'
    soup = bs(textPrice, 'html.parser')
    textPrice = soup.get_text()
    textPrice = textPrice.replace('[', '').replace(']', '')
    #<----------------------------------------------------------------->
    metaData.append(i)
    metaData.append(textTitle)
    metaData.append(textPrice)
    metaData.append(textRate)
    metaData.append(textNOR)
```

```python
metaData
```

```python
from openpyxl import Workbook
workbook = Workbook()
worksheet = workbook.active
def csvFile(name, price, rate, buyers ,url, numRow):
    worksheet.cell(row= numRow, column=1, value= name)
    worksheet.cell(row= numRow, column=2, value= price)
    worksheet.cell(row= numRow, column=3, value= rate)
    worksheet.cell(row= numRow, column=4, value= buyers)
    worksheet.cell(row= numRow, column=5, value= url)
    workbook.save('iphone.xlsx')
```

```python
i = 0
row = 1
while i <= size-5:
    link = metaData[i]
    textTitle = metaData[i+1]
    price = metaData[i+2]
    rating = metaData[i+3]
    buyers = metaData[i+4]
    csvFile(textTitle,price,rating,buyers,link, row)
    row += 1
    i += 5
```
